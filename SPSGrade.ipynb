```python
# Input the points earned for each grading component
midterm_1 = float(input("Please enter your Midterm 1 grade: "))
midterm_2 = float(input("Please enter your Midterm 2 grade: "))
essay_outline = float(input("Please enter your Essay Outline grade: "))
essay_final = float(input("Please enter your Essay Final Draft grade: "))
discussion = float(input("Please enter your Overall Discussion grade: "))
tophat = float(input("Please enter your Overall Top Hat grade: "))

# Define the percentages/weights
midterm_weight = 0.2
essay_outline_weight = 0.05
essay_final_weight = 0.25
discussion_weight = 0.2
tophat_weight = 0.1

# Calculate the grade
grade = (midterm_1 + midterm_2) * midterm_weight + essay_outline \
        * essay_outline_weight + essay_final * essay_final_weight \
        + discussion * discussion_weight + tophat * tophat_weight

# Select the letter grade
if grade >= 90:
  letter = "A"
elif grade >= 86:
  letter = "A-"
elif grade >= 82:
  letter = "B+"
elif grade >= 78:
  letter = "B"
elif grade >= 74:
  letter = "B-"
elif grade >= 70:
  letter = "C+"
elif grade >= 66:
  letter = "C"
elif grade >= 63:
  letter = "C-"
elif grade >= 58:
  letter = "D+"
elif grade >= 54:
  letter = "D"
else:
  letter = "F"

# Print the resulting Value
print("The overall grade is ", grade, " and the letter grade is ", letter, ".", sep="")
```
